```python
df = pd.read_csv('titanic.csv')
df.set_index('PassengerId', inplace=True)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# now I want to plot gender, as male and female
# to know that how many are male and female
df.Sex.value_counts().plot(kind = 'pie')
```

```python
#Now I will plot a scatter plot to know each that how much fares each gender had paid
sns.scatterplot(x = 'Age', y = 'Fare', hue = 'Sex', data = df)
```
